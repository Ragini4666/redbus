```python

import pandas as pd

d1 = pd.read_csv('andhra_state.csv')
d2 = pd.read_csv('kerala.csv')
d3 = pd.read_csv('telangana_state.csv')
d4 = pd.read_csv('kadamba.csv')
d5 = pd.read_csv('rajasthan.csv')
d6 = pd.read_csv('south_bengal.csv')
d7 = pd.read_csv('himachal.csv')
d8 = pd.read_csv('assam.csv')
d9 = pd.read_csv('uttarpradesh.csv')
d10 = pd.read_csv('westbengal.csv')
d11 = pd.read_csv('chandigarh.csv')
d12 = pd.read_csv('punjab.csv')


# List of CSV file paths
csv_files = ['andhra_state.csv', 'kerala.csv', 'telangana_state.csv', 'kadamba.csv', 
             'rajasthan.csv','south_bengal.csv','himachal.csv','assam.csv', 'uttarpradesh.csv', 
             'westbengal.csv', 'chandigarh.csv', 'punjab.csv']   

# Read each CSV file into a DataFrame and store them in a list
df_list = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Add an ID column to the DataFrame (1-based indexing)
combined_df.insert(0, 'Id', range(1, len(combined_df) + 1))

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("BUS_DETAILS.csv", index=False)

# Replace the 'INR ' prefix in the 'Price' column
if 'Price' in combined_df.columns:
    combined_df['Price'] = combined_df['Price'].str.replace('INR ', '', regex=False)


# Save the cleaned DataFrame to a new CSV file
combined_df.to_csv("Bus_routes.csv", index=False)
print("file saved")
```

```python
def classify_bus_type(Bus_Type):
    """
    Classifies bus types into detailed categories such as 
    'AC Seater', 'AC Sleeper', 'Non-AC Seater', and 'Non-AC Sleeper'.
    """
    # Convert to lowercase for case-insensitive matching
    Bus_Type = Bus_Type.lower()
    
    # Non-AC classification
    if 'non ac' in Bus_Type or 'non a/c' or 'non-ac' in Bus_Type:
        if 'sleeper' in Bus_Type:
            return 'Non-AC Sleeper'
        elif 'seater' in Bus_Type:
            return 'Non-AC Seater'
        else:
            return 'Non-AC'
    
    # AC classification
    elif 'a/c' in Bus_Type or 'ac' or 'a.c' in Bus_Type:
        if 'sleeper' in Bus_Type:
            return 'AC Sleeper'
        elif 'seater' in Bus_Type:
            return 'AC Seater'
        else:
            return 'AC'
    
    # Unknown type
    else:
        return 'Unknown'
        
# Apply the classification logic
read['AC_Category'] = read['Bus_Type'].apply(classify_bus_type)
```

```python
def classify_bus_name(Bus_Name):
    # Convert the bus name to lowercase for case-insensitive matching
    Bus_Name = Bus_Name.lower()
    
    # List of government bus keywords
    government_keywords = [
        'ksrtc', 'apsrtc', 'tgsrtc', 'ktcl', 'rsstc', 'sbstc', 
        'hrtc', 'astc', 'upsrtc', 'wntc', 'ctu', 'pepsu (punjab)'
    ]
    
    # Check if any keyword exists in the bus name
    if any(keyword in Bus_Name for keyword in government_keywords):
        return 'Government Bus'
    else:
        return 'Private Bus'

# Apply the classification logic
read['Bus_Category'] = read['Bus_Name'].apply(classify_bus_name)
```
