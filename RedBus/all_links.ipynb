```python

import pandas as pd

d1 = pd.read_csv("andhra_details.csv")
d2 = pd.read_csv("Kerala_details.csv")
d3 = pd.read_csv("telangana_route_details.csv")
d4 = pd.read_csv("kadamba_details.csv")
d5 = pd.read_csv("rsrtc_details.csv")
d6 = pd.read_csv("south_bengal_details.csv")
d7 = pd.read_csv("hrtc_details.csv")
d8 = pd.read_csv("assam_details.csv")
d9 = pd.read_csv("upsrtc_details.csv")
d10 = pd.read_csv("west_bengal_details.csv")
d11 = pd.read_csv("chandigarh_details.csv")
d12 = pd.read_csv("punjab_details.csv")


# List of CSV file paths
csv_files = ["andhra_details.csv", "Kerala_details.csv", "telangana_route_details.csv", "kadamba_details.csv", 
             "rsrtc_details.csv", "south_bengal_details.csv","hrtc_details.csv", "assam_details.csv", "upsrtc_details.csv", 
             "west_bengal_details.csv", "chandigarh_details.csv", "punjab_details.csv"]   

# Read each CSV file into a DataFrame and store them in a list
df_list = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Add an ID column to the DataFrame (1-based indexing)
combined_df.insert(0, 'Id', range(1, len(combined_df) + 1))

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("bus_routes_details.csv", index=False)

# Replace the 'INR ' prefix in the 'Price' column
if 'Price' in combined_df.columns:
    combined_df['Price'] = combined_df['Price'].str.replace('INR ', '', regex=False)


# Save the cleaned DataFrame to a new CSV file
combined_df.to_csv("bus_routes.csv", index=False)
print("file saved")
```

```python
#python mysql connection

myconnection = pymysql.connect(host='127.0.0.1', user='root',passwd='Ragini95@',database="redbus")
read = pd.read_csv("bus_routes.csv")
read = read.dropna() #drop null values
a = ",".join(f"{i} {j}"
for i,j in zip(read.columns,read.dtypes)).replace("float64","float").replace("object","text").replace("int64","int") #replace dtypes from pandas to mysql
table_name = "bus_routes"
myconnection.cursor().execute(f"create table {table_name} ({a})")
for i in range(len(read)):
    myconnection.cursor().execute(f"insert into {table_name} values {tuple(read.iloc[i])}")
    myconnection.commit()

print("data inserted successfully")
```
